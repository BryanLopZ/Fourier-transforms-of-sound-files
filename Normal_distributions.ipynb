```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import statistics
```

```python
# Plot between -10 and 10 with .001 steps.
def norm_dist(mean, sd):
    x_axis = np.arange(-3*sd+mean, 3*sd+mean, 0.01)

    # Calculating mean and standard deviation
    y = norm.pdf(x_axis, mean, sd)
    return y, x_axis
```

```python
name = ["a", "e","i", "o", "u", "m"]
Sd = [430.5294,175.3963,73.3293,108.797,100.564,80.85109]
means = [541.0361,447.9764,276.9061,442.6011,323.0901,236.5687]

for i in range(len(Sd)):
    y, x_axis = norm_dist(means[i], Sd[i])
    plt.plot(x_axis,y, label = name[i])

plt.title("Distribucin normal, primer armonico")
plt.legend()
plt.show()
    
```
